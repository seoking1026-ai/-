```python
# 셀 1: C 소스 파일 작성
%%writefile prac4_1.c
#include <pthread.h>
#include <stdio.h>
#include <stdlib.h>

void* runner(void *param);

int sum[4];

int main() {
    pthread_t tid[4];
    pthread_attr_t attr[4];
    void* status = NULL;

    for(int i=0; i<4; i++){
        pthread_attr_init(&attr[i]);
    }

    pthread_create(&tid[0], &attr[0], runner, "1");
    pthread_create(&tid[1], &attr[1], runner, "10001");
    pthread_create(&tid[2], &attr[2], runner, "20001");
    pthread_create(&tid[3], &attr[3], runner, "30001");

    for(int i=0; i<4; i++) {
        pthread_join(tid[i], &status);
    }

    int total = 0;
    for(int i=0; i<4; i++) {
        total += sum[i];
    }

    printf("total : %d\n\n", total);
    return 0;
}

void* runner(void* param) {
    int from = atoi(param);
    int to = from + 9999;

    int index = from/10000;
    for(int i=from; i<=to; i++){
        sum[index] += i;
    }
}
```

```python
### 교재 252p 복합문제 1번

# 셀 1: C 소스 파일 작성
%%writefile prac4_2.c
#include <pthread.h>
#include <stdio.h>
#include <stdlib.h>

// 조건 (2): 0으로 초기화된 전역 변수 선언
int sum[4] = {0, 0, 0, 0};

// 조건 (3): 스레드 함수 이름은 runner
void* runner(void* param) {
    int from = atoi(param);
    int to = from + 9999;
    int index = from / 10000;

    for(int i = from; i <= to; i++) {
        sum[index] += i;
    }
    return NULL;
}

int main() {
    // 조건 (1): pthread 라이브러리 사용
    pthread_t tid[4];
    pthread_attr_t attr[4];
    void* status = NULL;

    // thread 속성값 초기화
    for(int i = 0; i < 4; i++) {
        pthread_attr_init(&attr[i]);
    }

    // 조건 (4): pthread_create()로 4개의 스레드를 연속적으로 생성하여 동시에 실행
    pthread_create(&tid[0], &attr[0], runner, "1");
    pthread_create(&tid[1], &attr[1], runner, "10001");
    pthread_create(&tid[2], &attr[2], runner, "20001");
    pthread_create(&tid[3], &attr[3], runner, "30001");

    // 4개의 스레드가 모두 종료될 때까지 대기
    for(int i = 0; i < 4; i++) {
        pthread_join(tid[i], &status);
    }

    // 최종 합산 및 출력
    int total = 0;
    for(int i = 0; i < 4; i++) {
        total += sum[i];
    }

    printf("1에서 40000까지 4개의 스레드가 계산한 총 합은 %d\n", total);
    return 0;
}
```
